```python
import pandas as pd
daf=pd.read_csv('HousePrices.csv')
from pandas import DataFrame
from sklearn import linear_model
import tkinter as tk 
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
Price_House=daf
df = DataFrame(Price_House,columns=['Id','LotArea','TotalBsmtSF','GarageCars','Property_Sale_Price']) 

X = df[['LotArea']].astype(int) # here we have 2 input variables for multiple regression. If you just want to use one variable for simple linear regression, then use X = df['Interest_Rate'] for example.Alternatively, you may add additional variables within the brackets
Y = df['Property_Sale_Price'].astype(int) # output variable (what we are trying to predict)

# with sklearn
regr = linear_model.LinearRegression()
regr.fit(X, Y)

print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)

# tkinter GUI
root= tk.Tk()

canvas1 = tk.Canvas(root, width = 500, height = 300)
canvas1.pack()

# with sklearn
Intercept_result = ('Intercept: ', regr.intercept_)
label_Intercept = tk.Label(root, text=Intercept_result, justify = 'center')
canvas1.create_window(260, 220, window=label_Intercept)

# with sklearn
Coefficients_result  = ('Coefficients: ', regr.coef_)
label_Coefficients = tk.Label(root, text=Coefficients_result, justify = 'center')
canvas1.create_window(260, 240, window=label_Coefficients)


# New_Interest_Rate label and input box
label1 = tk.Label(root, text='Type LotArea: ')
canvas1.create_window(100, 100, window=label1)

entry1 = tk.Entry (root) # create 1st entry box
canvas1.create_window(270, 100, window=entry1)


def values(): 
    global New_LotArea #our 1st input variable
    New_LotArea = float(entry1.get()) 
    Prediction_result  = ('Predicted Price_House: ', regr.predict([[New_LotArea]]))
    label_Prediction = tk.Label(root, text= Prediction_result, bg='orange')
    canvas1.create_window(260, 280, window=label_Prediction)
    
button1 = tk.Button (root, text='Predict Price_House',command=values, bg='orange') # button to call the 'values' command above 
canvas1.create_window(270, 150, window=button1)
 



 
figure3 = plt.Figure(figsize=(5,4), dpi=100)
ax3 = figure3.add_subplot(111)
ax3.scatter(df['LotArea'].astype(float),df['Property_Sale_Price'].astype(float), color = 'r')
scatter3 = FigureCanvasTkAgg(figure3, root) 
scatter3.get_tk_widget().pack(side=tk.RIGHT, fill=tk.BOTH)
ax3.legend() 
ax3.set_xlabel('LotArea')
ax3.set_title('LotArea Vs. Property_Sale_Price')

 


root.mainloop()
```

```python
import pandas as pd
daf=pd.read_csv('HousePrices.csv')
from pandas import DataFrame
from sklearn import linear_model
import tkinter as tk 
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
House_Price=daf
df = DataFrame(House_Price,columns=['Id','LotArea','TotalBsmtSF','GarageCars','Property_Sale_Price']) 

X = df[['LotArea','TotalBsmtSF','GarageCars']].astype(int)
Y = df['Property_Sale_Price'].astype(int) 
regr = linear_model.LinearRegression()
regr.fit(X, Y)
print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)
root= tk.Tk()

canvas1 = tk.Canvas(root, width = 500, height = 300)
canvas1.pack()


Intercept_result = ('Intercept: ', regr.intercept_)
label_Intercept = tk.Label(root, text=Intercept_result, justify = 'center')
canvas1.create_window(260, 220, window=label_Intercept)


Coefficients_result  = ('Coefficients: ', regr.coef_)
label_Coefficients = tk.Label(root, text=Coefficients_result, justify = 'center')
canvas1.create_window(260, 240, window=label_Coefficients)



label1 = tk.Label(root, text='Type LotArea: ')
canvas1.create_window(100, 100, window=label1)

entry1 = tk.Entry (root) # create 1st entry box
canvas1.create_window(270, 100, window=entry1)


label2 = tk.Label(root, text=' Type TotalBsmtSF: ')
canvas1.create_window(120, 120, window=label2)

entry2 = tk.Entry (root) 
canvas1.create_window(270, 120, window=entry2)

label3 = tk.Label(root, text=' Type GarageCars: ')
canvas1.create_window(150, 150, window=label3)

entry3 = tk.Entry (root) 
canvas1.create_window(270, 150, window=entry3)

def values(): 
    global New_LotArea 
    New_LotArea = float(entry1.get()) 
    
    global New_TotalBsmtSF 
    New_TotalBsmtSF = float(entry2.get()) 
    
    global New_GarageCars 
    New_GarageCars = float(entry3.get()) 
    
    Prediction_result  = ('Predicted_Sale_Price: ', regr.predict([[ New_LotArea,New_TotalBsmtSF, New_GarageCars]]))
    label_Prediction = tk.Label(root, text= Prediction_result, bg='orange')
    canvas1.create_window(300, 300, window=label_Prediction)
    
button1 = tk.Button (root, text='Predicted_Sale_Price',command=values, bg='orange')
canvas1.create_window(300, 200, window=button1)
 







figure3 = plt.Figure(figsize=(5,4), dpi=100)
ax3 = figure3.add_subplot(111)
ax3.scatter(df['LotArea'].astype(float),df['Property_Sale_Price'].astype(float), color = 'r')
scatter3 = FigureCanvasTkAgg(figure3, root) 
scatter3.get_tk_widget().pack(side=tk.RIGHT, fill=tk.BOTH)
ax3.legend() 
ax3.set_xlabel('LotArea')
ax3.set_title('LotArea Vs. Property_Sale_Price')


figure4 = plt.Figure(figsize=(5,4), dpi=100)
ax4 = figure4.add_subplot(111)
ax4.scatter(df['TotalBsmtSF'].astype(float),df['Property_Sale_Price'].astype(float), color = 'g')
scatter4 = FigureCanvasTkAgg(figure4, root) 
scatter4.get_tk_widget().pack(side=tk.RIGHT, fill=tk.BOTH)
ax4.legend() 
ax4.set_xlabel('TotalBsmtSF')
ax4.set_title('TotalBsmtSF Vs. Property_Sale_Price')

figure4 = plt.Figure(figsize=(5,4), dpi=100)
ax4 = figure4.add_subplot(111)
ax4.scatter(df['GarageCars'].astype(float),df['Property_Sale_Price'].astype(float), color = 'y')
scatter4 = FigureCanvasTkAgg(figure4, root) 
scatter4.get_tk_widget().pack(side=tk.RIGHT, fill=tk.BOTH)
ax4.legend() 
ax4.set_xlabel('GarageCars')
ax4.set_title('GarageCars Vs. Property_Sale_Price')

root.mainloop()
```
